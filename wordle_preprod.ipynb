```python
palabras_crudas = [
    "altura", "blanco", "camino", "dormir", "espejo",
    "fuerza", "gritar", "hierro", "jardin", "lluvia",
    "ma√±ana", "nochez", "origen", "puerta", "quesos",
    "ratonz", "suerte", "tierra", "unidos", "viento"
]
numero_elegido = int(input('Ingresa un numero del 1 al 20 para elegir una palabra: '))
palabra_elegida = palabras_crudas[numero_elegido - 1]
print(palabra_elegida)


 
ingresar = input('Palabra aqui')

limite_letra = 6

def Resaltar_letra(palabra_ingresada,palabra_clave):
    letras_verificadas = []
    for a in range(limite_letra):
        letras_iguales = palabra_ingresada[a] == palabra_clave[a] 
    for l in palabra_ingresada:
            if letras_iguales: 
             letras_verificadas.append(f'[{l}]')
            elif l in palabra_clave:
             letras_verificadas.append(f'({l})')
            else:
              letras_verificadas.append(f'{l}')
    return letras_verificadas
les = Resaltar_letra(ingresar, palabra_elegida )
print(les)

 
```
